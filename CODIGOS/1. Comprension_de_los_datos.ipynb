## Verificación de la calidad de datos

### Visualización de datos nulos en las datas

```python
def verificar_porcentaje_nulos(prefix='data', cantidad=3):
    for i in range(1, cantidad + 1):
        nombre_var = f"{prefix}{i}"
        if nombre_var in globals():
            df = globals()[nombre_var]
            print(f"\n Porcentaje de nulos en {nombre_var}:")
            porcentaje_nulos = df.isnull().mean() * 100
            print(porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False).round(2))
            

         
```

```python
verificar_porcentaje_nulos(cantidad=3)
```

### Visualizar valores atípicos

```python
import math
def graficar_outliers(prefix='data', cantidad=3):
    for i in range(1, cantidad + 1):
        nombre_var = f"{prefix}{i}"
        if nombre_var in globals():
            df = globals()[nombre_var]
            columnas_numericas = df.select_dtypes(include='number')

            if columnas_numericas.empty:
                print(f"\n {nombre_var} no tiene columnas numéricas.")
                continue

            print(f"\n Collage de boxplots para {nombre_var}:")

            n = len(columnas_numericas.columns)
            cols = 3  # Número de columnas en el collage
            rows = math.ceil(n / cols)

            fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 3.5 * rows))
            axes = axes.flatten()

            for j, col in enumerate(columnas_numericas.columns):
                sns.boxplot(data=df, x=col, ax=axes[j], color='skyblue')
                axes[j].set_title(col)

            # Eliminar los ejes vacíos si hay menos variables que subplots
            for k in range(j+1, len(axes)):
                fig.delaxes(axes[k])

            fig.suptitle(f'Boxplots de variables numéricas en {nombre_var}', fontsize=16)
            plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajuste para el título general
            plt.show()
```

```python
graficar_outliers(cantidad=3)
```

```python
from rapidfuzz import process, fuzz

def sugerir_correcciones(df, umbral_similitud=85, columnas_excluidas=None):
    if columnas_excluidas is None:
        columnas_excluidas = ['CÓDIGO SINIESTRO', 'CÓDIGO VEHÍCULO', 'CÓDIGO PERSONA', 'EDAD','FECHA','CÓDIGO DE CARRETERA']

    columnas_objetivo = [
        col for col in df.select_dtypes(include='object').columns
        if col not in columnas_excluidas and df[col].isnull().sum() == 0
    ]

    for col in columnas_objetivo:
        print(f"\n🔍 Revisando columna: {col}")
        
        # Convertir todo a texto limpio (str, sin espacios, minúscula)
        valores = df[col].dropna().apply(lambda x: str(x).strip().lower()).unique()
        # Ordenar por frecuencia
        valores_ordenados = sorted(valores, key=lambda x: df[col].astype(str).str.strip().str.lower().tolist().count(x), reverse=True)

        sugerencias = []
        for i, valor in enumerate(valores_ordenados):
            candidatos = valores_ordenados[:i]
            resultado = process.extract(str(valor), [str(c) for c in candidatos], scorer=fuzz.ratio, limit=1)
            if resultado and resultado[0][1] >= umbral_similitud:
                sugerencias.append((valor, resultado[0][0], resultado[0][1]))

        if sugerencias:
            print("📌 Posibles errores ortográficos detectados:")
            for original, sugerido, similitud in sugerencias:
                print(f"  '{original}' ➡️ '{sugerido}' (similitud: {similitud}%)")
        else:
            print("✅ Sin errores evidentes detectados.")
```

```python
sugerir_correcciones(data1)
```

### Comprobación de valores

```python
# Convertir Edad a numérico y graficar distribución
data_edad = pd.to_numeric(data1['EDAD'], errors='coerce')  # evitar errores por strings mal escritos

# Ver estadísticas
print("📌 Estadísticas de EDAD:")
print(data_edad.describe())

# Filtrar edades extremas
edades_extremas = data_edad[(data_edad < 1) | (data_edad > 100)]
print(f"\n⚠️ Edades atípicas encontradas: {len(edades_extremas)} registros")
display(data_edad.drop_duplicates())

# Visualizar la distribución
plt.figure(figsize=(8, 4))
sns.histplot(data_edad, bins=30, kde=True)
plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python

horas_invalidas = data1[(data1['HORA'] < 0) | (data1['HORA'] > 23)]

print(f"\n⏰ Horas inválidas encontradas: {len(horas_invalidas)} registros")
display(horas_invalidas[['HORA']].drop_duplicates())

# Visualización rápida
plt.figure(figsize=(8, 4))
sns.histplot(data1['HORA'].dropna(), bins=24)
plt.title("Distribución de Hora")
plt.xlabel("Hora del día")
plt.ylabel("Frecuencia")
plt.show()
```

```python
variables = ['CANTIDAD DE FALLECIDOS', 'CANTIDAD DE LESIONADOS', 'CANTIDAD DE VEHICULOS DAÑADOS']

for var in variables:
    print(f"\n📈 Estadísticas para {var}:")
    print(data2[var].describe())

    # Detectar valores extremos mayores a percentil 99
    umbral_alto = data2[var].quantile(0.99)
    extremos = data2[data2[var] > umbral_alto]
    print(f"⚠️ Valores extremos > P99 ({umbral_alto:.2f}): {len(extremos)} casos")
    display(extremos[[var]].drop_duplicates())

    # Gráfico
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=data2[var])
    plt.title(f"{var}")
    plt.show()
```
