```python
from sklearn.ensemble import RandomForestClassifier

# Copia temporal
df_temp = df_final.copy()

# Filtrar registros con y sin el valor
df_known = df_temp[df_temp['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?'].notnull()]
df_unknown = df_temp[df_temp['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?'].isnull()]

# Variable objetivo para imputaci칩n
y_known = df_known['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?']

# Eliminar columnas no disponibles en df_unknown
X_known = df_known.drop(columns=['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?', 'C칍DIGO SINIESTRO', 'GRAVEDAD'])
X_unknown = df_unknown.drop(columns=['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?', 'C칍DIGO SINIESTRO', 'GRAVEDAD'])

# Codificar categ칩ricas (solo para ejemplo simple)
X_full = pd.get_dummies(pd.concat([X_known, X_unknown]), drop_first=True)
X_known_encoded = X_full.iloc[:len(X_known)]
X_unknown_encoded = X_full.iloc[len(X_known):]

# Entrenar modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_known_encoded, y_known)

from sklearn.impute import SimpleImputer

# Crear un imputador para reemplazar los NaNs con la moda (para categ칩ricas)
imputer = SimpleImputer(strategy='most_frequent')
X_unknown_encoded_imputed = pd.DataFrame(imputer.fit_transform(X_unknown_encoded), columns=X_unknown_encoded.columns)

# Ahora puedes predecir sin problema
imputed_values = rf.predict(X_unknown_encoded_imputed)

# Asignar los valores imputados
df_final.loc[df_final['쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?'].isnull(), '쯉E SOMETI칍 A DOSAJE ET칈LICO CUANTITATIVO?'] = imputed_values
```

```python
variables_5 = [
    'SUPERFICIE DE CALZADA',
    'PERFIL LONGITUDINAL V칈A',
    'CARACTER칈STICAS DE V칈A',
    'ZONIFICACI칍N',
    'CONDICI칍N CLIM츼TICA',
    'EXISTE CICLOV칈A'
]

for col in variables_5:
    moda = df_final[col].mode()[0]
    df_final[col].fillna(moda, inplace=True)
```

```python
# Calcular los cuartiles y el IQR
Q1 = df_final['EDAD'].quantile(0.25)
Q3 = df_final['EDAD'].quantile(0.75)
IQR = Q3 - Q1

# Definir l칤mites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Mostrar cu치ntos registros son considerados outliers
outliers = df_final[(df_final['EDAD'] < limite_inferior) | (df_final['EDAD'] > limite_superior)]
print(f"游뛂 Se eliminar치n {len(outliers)} registros con EDAD at칤pica.")

# Eliminar los outliers
df_final = df_final[(df_final['EDAD'] >= limite_inferior) & (df_final['EDAD'] <= limite_superior)]
```
