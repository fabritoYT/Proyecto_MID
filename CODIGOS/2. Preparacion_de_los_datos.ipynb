```python
from sklearn.ensemble import RandomForestClassifier

# Copia temporal
df_temp = df_final.copy()

# Filtrar registros con y sin el valor
df_known = df_temp[df_temp['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?'].notnull()]
df_unknown = df_temp[df_temp['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?'].isnull()]

# Variable objetivo para imputaci√≥n
y_known = df_known['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?']

# Eliminar columnas no disponibles en df_unknown
X_known = df_known.drop(columns=['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?', 'C√ìDIGO SINIESTRO', 'GRAVEDAD'])
X_unknown = df_unknown.drop(columns=['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?', 'C√ìDIGO SINIESTRO', 'GRAVEDAD'])

# Codificar categ√≥ricas (solo para ejemplo simple)
X_full = pd.get_dummies(pd.concat([X_known, X_unknown]), drop_first=True)
X_known_encoded = X_full.iloc[:len(X_known)]
X_unknown_encoded = X_full.iloc[len(X_known):]

# Entrenar modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_known_encoded, y_known)

from sklearn.impute import SimpleImputer

# Crear un imputador para reemplazar los NaNs con la moda (para categ√≥ricas)
imputer = SimpleImputer(strategy='most_frequent')
X_unknown_encoded_imputed = pd.DataFrame(imputer.fit_transform(X_unknown_encoded), columns=X_unknown_encoded.columns)

# Ahora puedes predecir sin problema
imputed_values = rf.predict(X_unknown_encoded_imputed)

# Asignar los valores imputados
df_final.loc[df_final['¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?'].isnull(), '¬øSE SOMETI√ì A DOSAJE ET√çLICO CUANTITATIVO?'] = imputed_values
```

```python
variables_5 = [
    'SUPERFICIE DE CALZADA',
    'PERFIL LONGITUDINAL V√çA',
    'CARACTER√çSTICAS DE V√çA',
    'ZONIFICACI√ìN',
    'CONDICI√ìN CLIM√ÅTICA',
    'EXISTE CICLOV√çA'
]

for col in variables_5:
    moda = df_final[col].mode()[0]
    df_final[col].fillna(moda, inplace=True)
```

```python
# Lista de columnas con menos del 2% de nulos a tratar
columnas_menos_2 = [
    'ESTADO MODALIDAD',
    'MODALIDAD DE TRANSPORTE',
    'ESTADO SOAT',
    'CAUSA',
    'TIPO DE V√çA',
    'HORA',
    'DIA',
    'VEH√çCULO',
    'ESTADO LICENCIA',
    'SEXO',
    'EDAD',
    'GRAVEDAD',
    'TIPO PERSONA',
    'DISTRITO'
]

# Eliminar registros con nulos en cualquiera de esas columnas
df_final = df_final.dropna(subset=columnas_menos_2).reset_index(drop=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Convertir a numerica
df_final['EDAD'] = pd.to_numeric(df_final['EDAD'], errors='coerce')

# Graficar boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_final['EDAD'])
plt.title('Valores at√≠picos en la variable EDAD')
plt.xlabel('Edad')
plt.grid(True)
plt.show()
```

```python
# Calcular los cuartiles y el IQR
Q1 = df_final['EDAD'].quantile(0.25)
Q3 = df_final['EDAD'].quantile(0.75)
IQR = Q3 - Q1

# Definir l√≠mites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Mostrar cu√°ntos registros son considerados outliers
outliers = df_final[(df_final['EDAD'] < limite_inferior) | (df_final['EDAD'] > limite_superior)]
print(f"üö´ Se eliminar√°n {len(outliers)} registros con EDAD at√≠pica.")

# Eliminar los outliers
df_final = df_final[(df_final['EDAD'] >= limite_inferior) & (df_final['EDAD'] <= limite_superior)]
```
